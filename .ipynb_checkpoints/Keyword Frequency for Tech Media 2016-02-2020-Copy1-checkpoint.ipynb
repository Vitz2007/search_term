```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn import metrics
%matplotlib inline
import os
os.getcwd()
```

```python
kw_freq = pd.read_csv('coefs_weighted_site.csv')
print(kw_freq.head())
```

```python
#Overview of data
print(kw_freq.info())
```

```python
#Find any NaN value
kw_freq.isnull().any().any()
print(kw_freq.isnull().sum())
```

```python
#Find any outliers
Q1 = kw_freq.quantile(0.25)
Q3 = kw_freq.quantile(0.75)
IQR = Q3-Q1
print(IQR)
```

```python
#Remove word 'freq' from year_month columns
kw_freq.columns = kw_freq.columns.str.replace(r'freq_', '')
print(kw_freq)
```

```python
#Separate punctuation from words in 'Unnamed: 0' column

def punctuation_marks(mark):
    out_punct = re.findall(r'[!"\$%&\'()*+,\-.\/:;=#@?\[\\\]^_`{|}~]*', mark)
    
    string = ''.join(out_punct)
    return list(string)

#Append to new column 'punctuation_mks'
kw_freq['punctuation_marks'] = kw_freq['Unnamed: 0'].apply(lambda x:punctuation_marks(x))
print(kw_freq)
```

```python
#Separate words from punctuation marks in column 'Unamed: 0' and put into new column 'words'
kw_freq['words'] = kw_freq['Unnamed: 0'].str.extract('(\w+)', expand=False)
print(kw_freq)
```

```python
#Check for new NaNs in 'words' column
print(kw_freq['words'].isnull().sum())
```

```python
#Replace NaNs with '0' from the newly created columns 'words'
kw_freq['words'] = kw_freq['words'].fillna(0)
print(kw_freq)
```

```python
#Convert [] to NaN and then replace with '0 values'
kw_freq = kw_freq.mask(kw_freq.applymap(type).eq(list) & ~kw_freq.astype(bool))
kw_freq['punctuation_marks'] = kw_freq['punctuation_marks'].fillna(0)
print(kw_freq)
```

```python
#Change 'punctuation_marks' column from 'object' to 'string type'
kw_freq['punctuation_marks'] = kw_freq['punctuation_marks'].astype(str)

#Remove brackets from 'punctuation_marks' column
kw_freq['punctuation_marks'] =  kw_freq['punctuation_marks'].apply(lambda x: x.replace('[','').replace(']',''))

#Remove quotes from values
kw_freq['punctuation_marks'] = kw_freq['punctuation_marks'].str.replace(r"[\"\']", '', regex=True)
print(kw_freq)
```

```python
#Move columns 'words' & 'punctuation_marks' to second and third position
#Pop the columns
col_name = 'words'
first_col = kw_freq.pop(col_name)
col_name2 = 'punctuation_marks'
second_col = kw_freq.pop(col_name2)
```

```python
#Insert columns back in dataframe
kw_freq.insert(1, col_name, first_col)
kw_freq.insert(2, col_name2, second_col)
print(kw_freq)
```

```python
#Drop 'Unnamed: 0' column now that we have split words and punctuation marks from it
kw_freq = kw_freq.drop('Unnamed: 0', 1)
```

```python
#Remove rows with stop words in 'words' column
stop = stopwords.words('english')
kw_freq['new_words'] = [True if row.words not in stop and 
                        row.punctuation_marks not in stop else False for index, row in kw_freq.iterrows()]
kw_freq = kw_freq[kw_freq.new_words == True]
kw_freq.drop('new_words', axis=1, inplace=True)
print(kw_freq.head(25))
```

```python
#Drop 'punctuation_marks', 'column to focus on 'words' frequencies
kw_freq = kw_freq.drop(columns=['punctuation_marks', 'coef', 'coef_norm_max', 'coef_norm'])
```

```python
#Remove rows with zero & last row
kw_freq = kw_freq[kw_freq['words']!=0]
kw_freq = kw_freq.drop([109927])
print(kw_freq)
```

```python
#Display 2000 rows of data
pd.options.display.max_rows = 1000

#Set 2500 rows of data as new dataframe for regression
kw_freq_lrg = kw_freq.head(1000)
print(kw_freq_lrg)
```

```python
#Convert 'word' colum from categorical to numerical column
labelencoder = LabelEncoder()
kw_freq_lrg['words'] = labelencoder.fit_transform(kw_freq_lrg['words'])
print(kw_freq_lrg['words'])
```

```python
#Separate data into features and target and save features
X = kw_freq_lrg.iloc[:, 1:]
y = kw_freq_lrg.iloc[:, 0]
```

```python
#Get shape of data
print(X.shape, y.shape)
```

```python
#Train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

```python
#Initialize Random Forest Regressor
fr = RandomForestRegressor(random_state=2)

#Use baseline without any parameter tuning
fr.fit(X_train, y_train)
```

```python
#Make prediction
y_pred = fr.predict(X_test)
print(y_pred)
```

```python
#Check accuracy of model
baseline_accu = fr.score(X_test, y_test)
print(baseline_accu)
```

```python
#Tuning parameters
n_estimators = [int(x) for x in np.linspace(start=100, stop=2000, num=10)]
max_depth = [int(x) for x in np.linspace(10, 110, num=10)]
max_depth.append(None)
min_samples_split = [2, 5, 10, 15, 25]
max_features = ['auto', 'sqrt', 'log2']
min_samples_leaf = [1, 2, 5, 10, 15]
bootstrap = [True, False]

#Create random grid
fr_param = {'n_estimators': n_estimators, 'max_depth': max_depth, 'min_samples_split': min_samples_split
            ,'max_features': max_features, 'bootstrap': bootstrap, 'min_samples_leaf': min_samples_leaf}
```

```python
#Initialize Random Forest Regressor with hyperparameters
fr = RandomForestRegressor()
fr_hyp = RandomizedSearchCV(estimator=fr, param_distributions = fr_param, cv=3, random_state=2,
                           verbose=2, n_jobs= -1, n_iter=300)
```

```python
#Fit random_parameter model
fr_hyp.fit(X_train, y_train)
```

```python
#Look at best parameters from fitted model
fr_hyp.best_params_
```

```python
#Make predictions with hyperparameters
fr_hyp.predict(X_test)
```

```python
#Check accuracy of model with hyperparameters
rand_search = fr_hyp.score(X_test, y_test)
print(rand_search)
```

```python
#Comparison accuracy from baseline model to hyperparameter model
def evaluate(model, X_test, y_test):
    predictions = model.predict(X_test)
    errors = abs(predictions - y_test)
    mape = 100 * np.mean(errors / y_test)
    accuracy = 100 - mape
    print('Model Performance')
    print('Average Error: {:0.4f} degrees.'.format(np.mean(errors)))
    print('Accuracy = {:0.2f}%.'.format(accuracy))
    
    return accuracy


fr_compare = RandomForestRegressor(random_state = 2)
fr_compare.fit(X_train, y_train)
baseline_accu = evaluate(fr_compare, X_test, y_test)


best_random = fr_hyp.best_estimator_
rand_search = evaluate(best_random, X_test, y_test)


print('Improvement of {:0.2f}%.'.format(100*(rand_search - baseline_accu) / baseline_accu))
```
